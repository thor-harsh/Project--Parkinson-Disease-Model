```python
import pandas as pd
```

```python
df=pd.read_csv('parkinson_disease.csv')
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.groupby('id').mean().reset_index()
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
df
```

```python
columns = list(df.columns)
for col in columns:
    if col == 'class':
        continue

    filtered_columns = [col]
    for col1 in df.columns:
        if((col == col1) | (col == 'class')):
            continue

        val = df[col].corr(df[col1])

        if val > 0.7:
        # If the correlation between the two
        # features is more than 0.7 remove
            columns.remove(col1)
            continue
        else:
            filtered_columns.append(col1)

# After each iteration filter out the columns
# which are not highly correlated features.
    df = df[filtered_columns]
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2
```

```python
X = df.drop('class', axis=1)
X_norm = MinMaxScaler().fit_transform(X)
selector = SelectKBest(chi2, k=30)
selector.fit(X_norm, df['class'])
filtered_columns = selector.get_support()
filtered_data = X.loc[:, filtered_columns]


# 
# 
```

```python
filtered_data['class'] = df['class']
df = filtered_data
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=df.drop('class',axis=1)
target=df['class']
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=10)
```

```python
X_train.shape,X_test.shape
```

```python
df.shape
```

```python
from sklearn.linear_model import LogisticRegression
# from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
X_train.head()
```

```python
X_train[:]
```

```python
from sklearn.metrics import roc_auc_score as ras
models=[LogisticRegression(max_iter=5000)]
for i in range(len(models)):
    models[i].fit(X_train,y_train)
    train_preds = models[i].predict_proba(X_train)[:,1]
    print('Training Accuracy : ', ras(y_train, train_preds))
 
    val_preds = models[i].predict_proba(X_test)[:,1]
    print('Validation Accuracy : ', ras(y_test, val_preds))
    print()
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report
```

```python
plot_confusion_matrix(models[0],X_test,y_test)
plt.show()
```

```python
print(classification_report(y_test,models[0].predict(X_test)))
```
